```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, json
```

```python
tmdb_2000 = pd.read_csv(r'C:\Users\dhowa\anaconda3\etc\profile.d\0130-Wk-13-local-workflow-practice\Project-3-Part-2\Data\final_tmdb_data2000.csv.gz')
tmdb_2000.info()
tmdb_2000.head()
```

```python
tmdb_2001 = pd.read_csv(r'C:\Users\dhowa\anaconda3\etc\profile.d\0130-Wk-13-local-workflow-practice\Project-3-Part-2\Data\final_tmdb_data2001.csv.gz')
tmdb_2001.info()
tmdb_2001.head()
```

```python
tmdb_combined_df = pd.concat((tmdb_2000, tmdb_2001), ignore_index=True)
tmdb_combined_df.info()
tmdb_combined_df.head()
```

```python
tmdb_combined_df.to_csv('Data/tmdb_results_combined.csv.gz')
```

```python
valid_finance_df = tmdb_combined_df[(tmdb_combined_df['budget'] > 0) | (tmdb_combined_df['revenue'] > 0)]
valid_finance_df.info()
valid_finance_df.head()
```

```python
valid_finance_df['certification'].value_counts().plot(kind='bar', 
                                                      color =['blue','red',
                                                             'purple', 'cyan',
                                                             'black'])
plt.title('Movie Counts By Certifications')
plt.xlabel('Certifications')
plt.ylabel('Counts')
plt.tight_layout;
```

```python
cert_counts = valid_finance_df['certification'].value_counts()
cert_counts
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
avg_rev = valid_finance_df.groupby('certification')['revenue'].mean()
avg_rev
```

```python
avg_rev.plot(kind='bar', color =['blue','orange', 'purple', 'pink','green'])
plt.title('Average Movie Revenues Per Certification')
plt.xlabel('Certifications')
plt.ylabel('Revenue in Dollars')
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout;
```

```python
avg_budg = valid_finance_df.groupby('certification')['budget'].mean()
avg_budg
```

```python
avg_budg.plot(kind='bar', color =['blue','red', 'orange', 'pink','green'])
plt.title('Average Movie Budget Per Certification')
plt.xlabel('Certifications')
plt.ylabel('Budget in Dollars')
plt.ticklabel_format(style='plain', axis='y')
plt.tight_layout;
```
